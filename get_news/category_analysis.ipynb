```python
import pandas as pd
```

```python
df = pd.read_csv('../data/SBER.csv')
df
```

```python
df['Type'].value_counts()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
type_counts_by_year = df.groupby(['Year', 'Type'])['Type'].count()
print(type_counts_by_year)
```

```python
# Подсчет значений Type для каждого года
type_counts_by_year = df.groupby(['Type', 'Year']).size().unstack(fill_value=0)

# Вывод результата
type_counts_by_year
```
